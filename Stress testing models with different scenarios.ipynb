```python
import random

def create_random_scenario():
  """Creates a random scenario for stress testing.

  Returns:
    A dictionary mapping scenario names to scenario values.
  """

  scenarios = {
    "unemployment_rate": random.uniform(0.05, 0.15),
    "gdp_growth": random.uniform(-0.05, 0.05),
    "inflation_rate": random.uniform(0.02, 0.08),
    "exchange_rate": random.uniform(0.95, 1.05),
    "economic_growth": random.uniform(-0.05, 0.05),
    "debt_to_gdp": random.uniform(0.5, 1.0),
    "30_year_treasury_rate": random.uniform(0.02, 0.08),
    "10_year_treasury_rate": random.uniform(0.02, 0.08),
    "5_year_treasury_rate": random.uniform(0.02, 0.08),
    "30_year_fixed_mortgage_rate": random.uniform(0.02, 0.08),
  }

  return scenarios

def create_model(scenarios):
  """Creates a model for stress testing.

  Args:
    scenarios: A dictionary mapping scenario names to scenario values.

  Returns:
    A model for stress testing.
  """

  model = {}

  for scenario_name, scenario_value in scenarios.items():
    model[scenario_name] = random.uniform(0.0, 1.0)

  return model

def score_model(model, scenarios):
  """Scores a model for stress testing.

  Args:
    model: A model for stress testing.
    scenarios: A dictionary mapping scenario names to scenario values.

  Returns:
    A score for the model.
  """

  score = 0.0

  for scenario_name, scenario_value in scenarios.items():
    score += model[scenario_name] * scenario_value

  return score

if __name__ == "__main__":
  scenarios = create_random_scenario()
  model = create_model(scenarios)
  score = score_model(model, scenarios)

  print(f"Score: {score}")
```
